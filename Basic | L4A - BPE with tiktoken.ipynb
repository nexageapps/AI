```python
# Author: Karthik Arjun
# LinkedIn: https://www.linkedin.com/in/karthik-arjun-a5b4a258/
# Comment: Beginner Level - Byte Pair Encoding [BPE] with tiktoken
# Reference: https://github.com/openai/tiktoken
# Creation Date: 16/11/2025
# Updated at : 16/11/2025

from importlib.metadata import version
import tiktoken
print("tiktoken version is ", version("tiktoken"))

tokenizer = tiktoken.get_encoding("gpt2")

text = (
    "Hello AI world! <|endoftext|>. Will go for coffee?"
)

integiers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integiers)


```
