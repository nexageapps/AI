```python
from IPython.lib.security import encode
# Author: Karthik Arjun
# LinkedIn: https://www.linkedin.com/in/karthik-arjun-a5b4a258/
# Comment: Beginner Level - Data sampling with a sliding window : Page 35
# Reference: Book Build a Large Language Model from Scracth: Sebastian Raschka
# Creation Date: 23/11/2025
# Updated at : 23/11/2025

#Task: Predict the next word

import torch
from torch.utils.data import Dataset, DataLoader

sentence = 'The cat sat on the wall'
tokens = sentence.split();
print(tokens)
vocab = {word : i for i , word in enumerate(tokens)}
print(vocab)
encoded = [vocab[word] for word in tokens]
print(encoded)

context_size = 4 #This is the sliding window size

class SlidingWindowDataset(Dataset):
  def __init__(self, encoded_tokens, context_size):
    self.data = []
    for i in range(len(encoded_tokens) - context_size):
      x = encoded_tokens[i: i + context_size]
      y = encoded_tokens[i+1 : i + context_size + 1]
      self.data.append((torch.tensor(x), torch.tensor(y)))
      print(self.data)
  def __len__(self):
    return len(self.data)



dataset = SlidingWindowDataset(encoded, context_size)
dataloader = DataLoader(dataset, batch_size=1, shuffle=False)

for x, y in dataloader:
  print("X: ", x.tolist() , "---> Y:" , y.tolist())
```
