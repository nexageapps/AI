## 1. Propositions

A **proposition** is a declarative statement that is either **True (T)** or **False (F)**.

| Statement | Proposition? | Why |
|---|---|---|
| "It is raining" | ✅ Yes | Can be T or F |
| "2 + 2 = 4" | ✅ Yes | Always T (tautology) |
| "Close the door" | ❌ No | Command, not a statement |
| "What time is it?" | ❌ No | Question |

We assign **letters** to propositions:
- **A** = "It is raining"
- **B** = "The sun is shining"
- **C** = "There is a rainbow"

```python
# Let's represent propositions as Python booleans
A = True   # It is raining
B = True   # The sun is shining
C = False  # There is a rainbow

print(f"A (It is raining):      {A}")
print(f"B (The sun is shining): {B}")
print(f"C (There is a rainbow): {C}")
```

## 2. Logical Connectives

We combine propositions using **connectives**:

| Symbol | Name | Python | Meaning |
|--------|------|--------|---------|
| ¬ | NOT (Negation) | `not` | Flips truth value |
| ∧ | AND (Conjunction) | `and` | Both must be true |
| ∨ | OR (Disjunction) | `or` | At least one true |
| → | IMPLIES (Conditional) | `not A or B` | If A then B |
| ↔ | BICONDITIONAL (Iff) | `A == B` | Same truth value |

```python
# Implementing logical connectives in Python

def NOT(p):
    """Negation: ¬p"""
    return not p

def AND(p, q):
    """Conjunction: p ∧ q"""
    return p and q

def OR(p, q):
    """Disjunction: p ∨ q"""
    return p or q

def IMPLIES(p, q):
    """Conditional: p → q (equivalent to ¬p ∨ q)"""
    return (not p) or q

def IFF(p, q):
    """Biconditional: p ↔ q"""
    return p == q

# Test with our propositions
print(f"¬A (NOT raining):     {NOT(A)}")
print(f"A ∧ B (raining AND sunny): {AND(A, B)}")
print(f"A ∨ C (raining OR rainbow): {OR(A, C)}")
print(f"A → C (raining IMPLIES rainbow): {IMPLIES(A, C)}")
print(f"A ↔ B (raining IFF sunny): {IFF(A, B)}")
```

## 3. The Rainbow Rule: (A ∧ B) → C

**"If it is raining AND the sun is shining, then there is a rainbow."**

This is a real-world example of a conditional (implication). Let's explore when this statement is true or false.

```python
def rainbow_rule(a, b, c):
    """(A ∧ B) → C: If raining AND sunny, then rainbow"""
    return IMPLIES(AND(a, b), c)

# Test different scenarios
scenarios = [
    (True,  True,  True,  "Raining + Sunny + Rainbow"),
    (True,  True,  False, "Raining + Sunny + No Rainbow"),
    (True,  False, True,  "Raining + Cloudy + Rainbow"),
    (True,  False, False, "Raining + Cloudy + No Rainbow"),
    (False, True,  True,  "Dry + Sunny + Rainbow"),
    (False, True,  False, "Dry + Sunny + No Rainbow"),
    (False, False, True,  "Dry + Cloudy + Rainbow"),
    (False, False, False, "Dry + Cloudy + No Rainbow"),
]

print(f"{'Scenario':<35} {'A':>3} {'B':>3} {'C':>3} {'(A∧B)→C':>8}")
print("-" * 58)
for a, b, c, desc in scenarios:
    result = rainbow_rule(a, b, c)
    print(f"{desc:<35} {'T' if a else 'F':>3} {'T' if b else 'F':>3} {'T' if c else 'F':>3} {'T' if result else 'F':>8}")

print("\n💡 Key insight: The rule is only FALSE when it's raining AND sunny but there's NO rainbow.")
print("   An implication is false only when the premise is true and the conclusion is false.")
```

## 4. Truth Table Generator

A truth table shows the result of a logical expression for **every possible combination** of truth values.

For *n* propositions, there are **2ⁿ** rows.

```python
from itertools import product

def truth_table(variables, expression_fn, expression_str):
    """Generate and display a truth table."""
    n = len(variables)
    
    # Header
    header = "  ".join(f"{v:>3}" for v in variables) + f"  {expression_str:>15}"
    print(header)
    print("-" * len(header))
    
    # All combinations
    for values in product([True, False], repeat=n):
        assignment = dict(zip(variables, values))
        result = expression_fn(**assignment)
        row = "  ".join(f"{'T' if v else 'F':>3}" for v in values)
        row += f"  {'T' if result else 'F':>15}"
        print(row)

# Truth table for (A ∧ B) → C
print("Truth Table: (A ∧ B) → C")
print("=" * 40)
truth_table(
    ['A', 'B', 'C'],
    lambda A, B, C: IMPLIES(AND(A, B), C),
    '(A∧B)→C'
)

print("\n")

# Truth table for P → Q
print("Truth Table: P → Q")
print("=" * 30)
truth_table(
    ['P', 'Q'],
    lambda P, Q: IMPLIES(P, Q),
    'P→Q'
)
```

## 5. Important Logical Equivalences

Two expressions are **logically equivalent** (≡) if they have the same truth table.

```python
def verify_equivalence(variables, expr1_fn, expr2_fn, name1, name2):
    """Verify two expressions are logically equivalent."""
    n = len(variables)
    all_match = True
    
    print(f"Checking: {name1} ≡ {name2}")
    header = "  ".join(f"{v:>3}" for v in variables) + f"  {name1:>12}  {name2:>12}  {'Match':>6}"
    print(header)
    print("-" * len(header))
    
    for values in product([True, False], repeat=n):
        assignment = dict(zip(variables, values))
        r1 = expr1_fn(**assignment)
        r2 = expr2_fn(**assignment)
        match = r1 == r2
        if not match:
            all_match = False
        row = "  ".join(f"{'T' if v else 'F':>3}" for v in values)
        row += f"  {'T' if r1 else 'F':>12}  {'T' if r2 else 'F':>12}  {'✓' if match else '✗':>6}"
        print(row)
    
    print(f"\n{'✅ EQUIVALENT' if all_match else '❌ NOT EQUIVALENT'}\n")

# De Morgan's Law: ¬(A ∧ B) ≡ (¬A ∨ ¬B)
verify_equivalence(
    ['A', 'B'],
    lambda A, B: NOT(AND(A, B)),
    lambda A, B: OR(NOT(A), NOT(B)),
    '¬(A∧B)', '(¬A∨¬B)'
)

# Contrapositive: (A → B) ≡ (¬B → ¬A)
verify_equivalence(
    ['A', 'B'],
    lambda A, B: IMPLIES(A, B),
    lambda A, B: IMPLIES(NOT(B), NOT(A)),
    'A→B', '¬B→¬A'
)

# Material Implication: (A → B) ≡ (¬A ∨ B)
verify_equivalence(
    ['A', 'B'],
    lambda A, B: IMPLIES(A, B),
    lambda A, B: OR(NOT(A), B),
    'A→B', '¬A∨B'
)
```

## 6. Inference Rules

Inference rules let us derive new truths from existing ones.

### Modus Ponens
If **P → Q** and **P**, then **Q**.

### Modus Tollens
If **P → Q** and **¬Q**, then **¬P**.

```python
def modus_ponens(p_implies_q, p):
    """If P→Q is true and P is true, conclude Q."""
    if p_implies_q and p:
        return True  # Q must be true
    return None  # Cannot conclude

def modus_tollens(p_implies_q, not_q):
    """If P→Q is true and ¬Q is true, conclude ¬P."""
    if p_implies_q and not_q:
        return True  # ¬P must be true
    return None  # Cannot conclude

# Example: "If it rains, the ground is wet"
print("=== Modus Ponens ===")
print("Rule: If it rains (P), the ground is wet (Q)")
print("Given: It is raining (P = True)")
print("Given: P → Q is True")
result = modus_ponens(True, True)
print(f"Conclusion: The ground is wet (Q) = {result}")

print("\n=== Modus Tollens ===")
print("Rule: If it rains (P), the ground is wet (Q)")
print("Given: The ground is NOT wet (¬Q = True)")
print("Given: P → Q is True")
result = modus_tollens(True, True)
print(f"Conclusion: It is NOT raining (¬P) = {result}")
```

## 7. Connection to Neural Networks

A single **perceptron** (the simplest neural network unit) can compute basic logical operations.

This is the bridge between symbolic logic and machine learning!

```python
import numpy as np

def perceptron(inputs, weights, bias):
    """Simple perceptron: output = step(w·x + b)"""
    return int(np.dot(inputs, weights) + bias >= 0)

# AND gate as a perceptron
print("AND gate (perceptron: w=[1,1], b=-1.5)")
print("A  B  | A ∧ B | Perceptron")
print("-" * 30)
for a, b in product([0, 1], repeat=2):
    logic_result = int(a and b)
    nn_result = perceptron([a, b], [1, 1], -1.5)
    print(f"{a}  {b}  |   {logic_result}   |     {nn_result}")

print("\nOR gate (perceptron: w=[1,1], b=-0.5)")
print("A  B  | A ∨ B | Perceptron")
print("-" * 30)
for a, b in product([0, 1], repeat=2):
    logic_result = int(a or b)
    nn_result = perceptron([a, b], [1, 1], -0.5)
    print(f"{a}  {b}  |   {logic_result}   |     {nn_result}")

print("\nNOT gate (perceptron: w=[-1], b=0.5)")
print("A  | ¬A | Perceptron")
print("-" * 22)
for a in [0, 1]:
    logic_result = int(not a)
    nn_result = perceptron([a], [-1], 0.5)
    print(f"{a}  |  {logic_result} |     {nn_result}")

print("\n💡 Every logical connective can be computed by a neural network!")
print("   This is why logic is foundational to understanding AI.")
```

## 9. First-Order Logic (FOL)

Propositional logic has a big limitation: it can only talk about specific facts. It can't express "**all** cats have whiskers" or "**some** dog is friendly." First-Order Logic (FOL) extends propositional logic with:

| Concept | Symbol | Example |
|---|---|---|
| Variables | x, y, z | Represent objects |
| Predicates | P(x), Loves(x,y) | Properties and relations |
| Universal quantifier | ∀x | "For all x" |
| Existential quantifier | ∃x | "There exists an x" |
| Constants | Alice, Bob | Specific objects |

**Examples:**
- `∀x Cat(x) → HasWhiskers(x)` — "All cats have whiskers"
- `∃x Dog(x) ∧ Friendly(x)` — "There exists a friendly dog"
- `∀x ∀y Parent(x,y) → Loves(x,y)` — "All parents love their children"

```python
# First-Order Logic in Python
# We represent a "model" as a domain + predicate interpretations

class FOLModel:
    """A simple First-Order Logic model."""
    def __init__(self, domain, predicates):
        self.domain = domain
        self.predicates = predicates  # dict of predicate_name -> set of tuples
    
    def holds(self, predicate, *args):
        """Check if predicate(args) is true in this model."""
        return args in self.predicates.get(predicate, set())
    
    def forall(self, predicate_fn):
        """∀x predicate_fn(x) — true if predicate holds for ALL domain elements."""
        return all(predicate_fn(x) for x in self.domain)
    
    def exists(self, predicate_fn):
        """∃x predicate_fn(x) — true if predicate holds for SOME domain element."""
        return any(predicate_fn(x) for x in self.domain)

# Create a small world
animals = FOLModel(
    domain={'Whiskers', 'Rex', 'Tweety', 'Goldie'},
    predicates={
        'Cat':    {('Whiskers',)},
        'Dog':    {('Rex',)},
        'Bird':   {('Tweety',)},
        'Fish':   {('Goldie',)},
        'Furry':  {('Whiskers',), ('Rex',)},
        'CanFly': {('Tweety',)},
        'Friendly': {('Rex',), ('Goldie',)},
    }
)

# ∀x Cat(x) → Furry(x)  "All cats are furry"
result = animals.forall(lambda x: not animals.holds('Cat', x) or animals.holds('Furry', x))
print(f"∀x Cat(x) → Furry(x): {result}  (All cats are furry)")

# ∃x Dog(x) ∧ Friendly(x)  "There exists a friendly dog"
result = animals.exists(lambda x: animals.holds('Dog', x) and animals.holds('Friendly', x))
print(f"∃x Dog(x) ∧ Friendly(x): {result}  (Some dog is friendly)")

# ∀x Bird(x) → CanFly(x)  "All birds can fly"
result = animals.forall(lambda x: not animals.holds('Bird', x) or animals.holds('CanFly', x))
print(f"∀x Bird(x) → CanFly(x): {result}  (All birds can fly)")

# ∃x CanFly(x) ∧ Fish(x)  "There exists a flying fish"
result = animals.exists(lambda x: animals.holds('CanFly', x) and animals.holds('Fish', x))
print(f"∃x CanFly(x) ∧ Fish(x): {result}  (Some fish can fly)")
```

## 10. Free Variables and Sentences

A **bound variable** is inside a quantifier (∀ or ∃). A **free variable** is not.

A **sentence** is a formula with **no free variables** — only sentences can be evaluated as TRUE or FALSE.

| Formula | Free Variables | Sentence? |
|---|---|---|
| `∀x Likes(x, IceCream)` | None | ✅ Yes |
| `Likes(x, y)` | x, y | ❌ No |
| `∀x ∃y Loves(x, y)` | None | ✅ Yes |
| `∀x Loves(x, y)` | y | ❌ No |

```python
# Demonstrating free vs bound variables

def analyze_formula(formula, quantified_vars, all_vars):
    """Identify free and bound variables in a formula."""
    free = [v for v in all_vars if v not in quantified_vars]
    bound = [v for v in all_vars if v in quantified_vars]
    is_sentence = len(free) == 0
    
    print(f"Formula: {formula}")
    print(f"  Bound variables: {bound if bound else 'None'}")
    print(f"  Free variables:  {free if free else 'None'}")
    print(f"  Is a sentence?   {'✅ Yes' if is_sentence else '❌ No'}")
    print()

analyze_formula('∀x Likes(x, IceCream)', quantified_vars={'x'}, all_vars=['x'])
analyze_formula('Likes(x, y)', quantified_vars=set(), all_vars=['x', 'y'])
analyze_formula('∀x ∃y Loves(x, y)', quantified_vars={'x', 'y'}, all_vars=['x', 'y'])
analyze_formula('∀x Loves(x, y)', quantified_vars={'x'}, all_vars=['x', 'y'])
```

## 11. Satisfaction Relation and Models

A **model** M gives meaning to our logic by defining:
- A **domain** (set of objects)
- **Interpretations** for predicates and constants

We say **M ⊨ φ** ("M satisfies φ") if sentence φ is TRUE in model M.

A sentence is:
- **Valid** (tautology) — true in ALL models
- **Satisfiable** — true in SOME model
- **Unsatisfiable** — true in NO model

```python
# Satisfaction: checking if a model satisfies a sentence

family = FOLModel(
    domain={'Alice', 'Bob', 'Charlie'},
    predicates={
        'Parent': {('Alice', 'Bob'), ('Bob', 'Charlie')},
        'Female': {('Alice',)},
        'Male':   {('Bob',), ('Charlie',)},
    }
)

# M ⊨ ∃x ∃y Parent(x,y)  "Someone is a parent of someone"
result = family.exists(lambda x: family.exists(lambda y: family.holds('Parent', x, y)))
print(f"M ⊨ ∃x ∃y Parent(x,y): {result}")

# M ⊨ ∀x ∃y Parent(x,y)  "Everyone is a parent of someone"
result = family.forall(lambda x: family.exists(lambda y: family.holds('Parent', x, y)))
print(f"M ⊨ ∀x ∃y Parent(x,y): {result}  (Charlie has no child!)")

# M ⊨ ∃x Female(x) ∧ ∃y Parent(x,y)  "Some female is a parent"
result = family.exists(lambda x: family.holds('Female', x) and 
                       family.exists(lambda y: family.holds('Parent', x, y)))
print(f"M ⊨ ∃x Female(x) ∧ Parent(x,_): {result}  (Alice is a mother)")
```

## 12. FOL Logical Equivalence

Key equivalences involving quantifiers:

| Equivalence | Meaning |
|---|---|
| `¬∀x P(x) ≡ ∃x ¬P(x)` | "Not all are P" = "Some are not P" |
| `¬∃x P(x) ≡ ∀x ¬P(x)` | "None are P" = "All are not P" |
| `∀x (P(x) ∧ Q(x)) ≡ (∀x P(x)) ∧ (∀x Q(x))` | ∀ distributes over ∧ |
| `∃x (P(x) ∨ Q(x)) ≡ (∃x P(x)) ∨ (∃x Q(x))` | ∃ distributes over ∨ |

⚠️ **Careful!** `∀x ∃y P(x,y)` is NOT the same as `∃y ∀x P(x,y)`
- "Everyone loves someone" ≠ "Someone is loved by everyone"

```python
# Verifying FOL equivalences

# ¬∀x P(x) ≡ ∃x ¬P(x)
# Using our animals model with 'Furry' predicate
lhs = not animals.forall(lambda x: animals.holds('Furry', x))
rhs = animals.exists(lambda x: not animals.holds('Furry', x))
print(f"¬∀x Furry(x) = {lhs}")
print(f"∃x ¬Furry(x) = {rhs}")
print(f"Equivalent? {'✅ Yes' if lhs == rhs else '❌ No'}")
print()

# ¬∃x P(x) ≡ ∀x ¬P(x)
lhs = not animals.exists(lambda x: animals.holds('Fish', x) and animals.holds('CanFly', x))
rhs = animals.forall(lambda x: not (animals.holds('Fish', x) and animals.holds('CanFly', x)))
print(f"¬∃x (Fish(x) ∧ CanFly(x)) = {lhs}")
print(f"∀x ¬(Fish(x) ∧ CanFly(x)) = {rhs}")
print(f"Equivalent? {'✅ Yes' if lhs == rhs else '❌ No'}")
print()

# ⚠️ ∀x ∃y ≠ ∃y ∀x (order matters!)
loves = FOLModel(
    domain={'A', 'B', 'C'},
    predicates={'Loves': {('A','B'), ('B','C'), ('C','A')}}
)
# ∀x ∃y Loves(x,y) — everyone loves someone
r1 = loves.forall(lambda x: loves.exists(lambda y: loves.holds('Loves', x, y)))
# ∃y ∀x Loves(x,y) — someone is loved by everyone
r2 = loves.exists(lambda y: loves.forall(lambda x: loves.holds('Loves', x, y)))
print(f"∀x ∃y Loves(x,y) = {r1}  (everyone loves someone)")
print(f"∃y ∀x Loves(x,y) = {r2}  (someone is loved by all)")
print(f"Same? {'✅ Yes' if r1 == r2 else '⚠️ No — order of quantifiers matters!'}")
```

## 13. FOL with Kinship Domain

Kinship (family relationships) is a classic FOL domain used in AI. We can define complex relations from simple ones:

| Defined Relation | FOL Definition |
|---|---|
| Mother(x,y) | Female(x) ∧ Parent(x,y) |
| Father(x,y) | Male(x) ∧ Parent(x,y) |
| Grandparent(x,y) | ∃z Parent(x,z) ∧ Parent(z,y) |
| Sibling(x,y) | x≠y ∧ ∃z Parent(z,x) ∧ Parent(z,y) |
| Ancestor(x,y) | Parent(x,y) ∨ ∃z Parent(x,z) ∧ Ancestor(z,y) |

```python
# Kinship domain in FOL

kinship = FOLModel(
    domain={'Alice', 'Bob', 'Charlie', 'Diana', 'Eve'},
    predicates={
        'Parent': {('Alice','Bob'), ('Alice','Charlie'), ('Bob','Diana'), ('Bob','Eve')},
        'Female': {('Alice',), ('Charlie',), ('Diana',), ('Eve',)},
        'Male':   {('Bob',)},
    }
)

def is_mother(model, x, y):
    return model.holds('Female', x) and model.holds('Parent', x, y)

def is_father(model, x, y):
    return model.holds('Male', x) and model.holds('Parent', x, y)

def is_grandparent(model, x, y):
    return any(model.holds('Parent', x, z) and model.holds('Parent', z, y) 
               for z in model.domain)

def is_sibling(model, x, y):
    return x != y and any(model.holds('Parent', z, x) and model.holds('Parent', z, y) 
                          for z in model.domain)

# Query the kinship model
print("=== Kinship Queries ===")
print(f"Mother(Alice, Bob)?      {is_mother(kinship, 'Alice', 'Bob')}")
print(f"Father(Bob, Diana)?      {is_father(kinship, 'Bob', 'Diana')}")
print(f"Grandparent(Alice, Diana)? {is_grandparent(kinship, 'Alice', 'Diana')}")
print(f"Sibling(Bob, Charlie)?   {is_sibling(kinship, 'Bob', 'Charlie')}")
print(f"Sibling(Diana, Eve)?     {is_sibling(kinship, 'Diana', 'Eve')}")

# Find all mothers
print("\n=== All Mothers ===")
for x in kinship.domain:
    for y in kinship.domain:
        if is_mother(kinship, x, y):
            print(f"  {x} is mother of {y}")

# Find all grandparents
print("\n=== All Grandparents ===")
for x in kinship.domain:
    for y in kinship.domain:
        if is_grandparent(kinship, x, y):
            print(f"  {x} is grandparent of {y}")

# Find all siblings
print("\n=== All Siblings ===")
seen = set()
for x in kinship.domain:
    for y in kinship.domain:
        pair = tuple(sorted([x, y]))
        if is_sibling(kinship, x, y) and pair not in seen:
            seen.add(pair)
            print(f"  {x} and {y} are siblings")

print("\n💡 This is exactly how AI knowledge bases work!")
print("   The Wumpus World agent uses similar FOL reasoning.")
```
