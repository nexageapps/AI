## Key Takeaways

**Relevant UoA Courses:** COMPSCI 703, COMPSCI 714

1. Language model predicts next token: P(w_t | w_1, ..., w_{t-1})
2. GPT architecture: decoder-only transformer with causal masking
3. Causal masking prevents attending to future tokens
4. Training: predict next token given previous tokens (self-supervised)
5. Generation: sample from output distribution autoregressively

## Exam Preparation Guide

### Essential Concepts for Exams

- Understand autoregressive generation: use previous predictions as input
- Know difference between encoder-only (BERT), decoder-only (GPT), encoder-decoder (T5)
- Explain causal masking: why and how it's implemented
- Calculate perplexity: exp(average negative log-likelihood)
- Common question: Describe GPT training objective and architecture

### Common Mistakes to Avoid

- ❌ Confusing causal (GPT) with bidirectional (BERT) attention
- ❌ Not understanding autoregressive generation process
- ❌ Forgetting that language models are trained with self-supervision
