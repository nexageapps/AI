## Exam Preparation Guide

### Essential Concepts for Exams

- Know how to approach new problems systematically
- Understand importance of baselines (random, simple models)
- Be able to debug common issues: NaN loss, exploding gradients, poor convergence
- Explain experimental design: train/val/test split, cross-validation
- Common question: Given a problem, outline your approach

### Common Mistakes to Avoid

- ❌ Starting with complex models before trying simple baselines
- ❌ Not tracking experiments and hyperparameters
- ❌ Overfitting to validation set by tuning too much

### Practice Problems

1. Design experiment to compare 3 models fairly
2. Loss becomes NaN during training. What could be wrong?
3. Outline approach for image classification project
