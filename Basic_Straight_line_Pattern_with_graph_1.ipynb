```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Training data
X = np.array([1, 2, 3, 4], dtype=float)
Y = np.array([27, 47, 67, 87], dtype=float) # y = 20x + 7

# Build a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Train the model
model.fit(X, Y, epochs=500, verbose=0)

# See what it learned
weights, biases = model.layers[0].get_weights()
print("Weight:", weights)  # Should be ~20
print("Bias:", biases)     # Should be ~7

# Convert arrays to scalars and round or convert to int
w = int(round(weights[0][0]))
b = int(round(biases[0]))

# Predict values
X_test = np.array([10,20,30,40,50])
Y_pred = model.predict(X_test)

# Plot original data and predicted line
plt.scatter(X, Y, color='blue', label='Training Data')
plt.scatter(X_test, Y_pred, color='red', label='Model Prediction')
plt.title("Model Prediction vs Actual Data")
label_text = f"Input (X) - The Prediction formula is: y = {w}x + {b}"
plt.xlabel(label_text)
plt.ylabel("Output (Y)")
plt.legend()
plt.grid(True)
plt.show()
```
