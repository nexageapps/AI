## Part 1: The Three-Pass Method for Reading Papers

### Pass 1: The Quick Scan (5-10 minutes)
**Goal:** Decide if the paper is worth reading in detail

**What to read:**
- Title and abstract
- Introduction and conclusion
- Section and subsection headings
- Figures and captions
- References (check if you know the cited work)

**Questions to answer:**
1. What is the main contribution?
2. Is it relevant to my work?
3. Is it well-written and rigorous?
4. What are the key results?

### Pass 2: Detailed Reading (1 hour)
**Goal:** Understand the paper's content but not the details

**What to do:**
- Read carefully but skip proofs
- Note key points and insights
- Mark unclear sections
- Study figures and tables thoroughly
- Check references for background

**Questions to answer:**
1. What problem does it solve?
2. What is the proposed method?
3. What are the assumptions?
4. How does it compare to prior work?
5. What are the experimental results?

### Pass 3: Deep Understanding (4-5 hours)
**Goal:** Virtually re-implement the paper in your mind

**What to do:**
- Understand every equation
- Work through proofs
- Question every assumption
- Think about alternative approaches
- Identify strengths and weaknesses

**Questions to answer:**
1. Can I explain this to someone else?
2. What are the limitations?
3. How would I improve it?
4. Can I reproduce the results?

## Exam Preparation Guide

### Essential Concepts for Exams

- Know how to analyze papers: novelty, experimental rigor, reproducibility
- Understand common experimental flaws: cherry-picking, unfair comparisons
- Be able to critique methodology and results
- Explain how to reproduce paper results
- Common question: Given paper abstract, identify key contribution

### Common Mistakes to Avoid

- ❌ Accepting results without checking experimental setup
- ❌ Not reading related work to understand context
- ❌ Focusing only on results, ignoring methodology

### Practice Problems

1. Read paper abstract, identify: problem, method, key result
2. What makes a good baseline in ML experiments?
3. How would you verify paper's claims?

### How This Helps Your UoA Courses

**COMPSCI 761, COMPSCI 703:**
- Provides hands-on implementation of theoretical concepts
- Practice problems similar to exam questions

### Study Tips for Advanced Topics

1. **Connect to Fundamentals**: Link advanced concepts to basics
2. **Read Papers**: Understand state-of-the-art approaches
3. **Implement from Scratch**: Deepens understanding
4. **Compare Approaches**: Know trade-offs between methods
5. **Real-World Applications**: Understand practical use cases

### Exam Question Types

- **Conceptual**: Explain advanced mechanisms and why they work
- **Comparison**: Compare multiple approaches, trade-offs
- **Design**: Design system for specific requirements
- **Analysis**: Analyze experimental results, identify issues
- **Application**: Apply techniques to novel problems
