**Complete an original research project and contribute to the field**

---

## Learning Objectives

- Identify research gaps and opportunities
- Design novel solutions
- Implement and evaluate rigorously
- Write and submit research papers
- Contribute to open-source research
- Build research portfolio

**Duration:** 4-8 weeks  
**Prerequisites:** E01-E14

---

## Project Guidelines

### Phase 1: Research Question (Week 1)
- Literature review
- Identify gaps
- Formulate hypothesis
- Define scope

### Phase 2: Design (Week 2)
- Propose solution
- Design experiments
- Plan evaluation
- Get feedback

### Phase 3: Implementation (Weeks 3-5)
- Build baseline
- Implement method
- Run experiments
- Iterate and refine

### Phase 4: Evaluation (Week 6)
- Comprehensive testing
- Ablation studies
- Statistical analysis
- Compare to baselines

### Phase 5: Writing (Week 7)
- Draft paper
- Create visualizations
- Write documentation
- Prepare code release

### Phase 6: Submission (Week 8)
- Final revisions
- Submit to conference/journal
- Release code on GitHub
- Share on social media
