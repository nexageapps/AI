## 1. Hyperparameter Tuning Fundamentals

### What are Hyperparameters?

**Model Hyperparameters:**
- Number of layers
- Number of units per layer
- Activation functions
- Architecture choices

**Training Hyperparameters:**
- Learning rate
- Batch size
- Number of epochs
- Optimizer choice

**Regularization Hyperparameters:**
- Dropout rate
- L1/L2 regularization strength
- Data augmentation parameters

### Search Strategies

1. **Manual Search:** Trial and error (slow, inefficient)
2. **Grid Search:** Exhaustive search (thorough but expensive)
3. **Random Search:** Random sampling (faster, often better)
4. **Bayesian Optimization:** Smart search (most efficient)
5. **Evolutionary Algorithms:** Population-based (for NAS)
