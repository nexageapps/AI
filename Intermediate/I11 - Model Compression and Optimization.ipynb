## 1. Why Model Compression?

### The Problem

**Large models:**
- High memory requirements
- Slow inference
- High energy consumption
- Can't run on mobile/edge devices

**Example:**
- GPT-3: 175B parameters, 700GB storage
- BERT-Large: 340M parameters, 1.3GB storage
- ResNet-50: 25M parameters, 100MB storage

### Compression Techniques

1. **Pruning:** Remove unnecessary weights
2. **Quantization:** Reduce precision (FP32 → INT8)
3. **Knowledge Distillation:** Train small model from large
4. **Low-rank Factorization:** Decompose weight matrices
5. **Weight Sharing:** Share weights across layers
