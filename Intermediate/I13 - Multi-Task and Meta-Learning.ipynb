## 1. Multi-Task Learning

### Concept

**Single-Task Learning:**
- One model, one task
- Example: Image classification only

**Multi-Task Learning:**
- One model, multiple related tasks
- Shared representations
- Example: Classification + segmentation + detection

### Benefits

1. **Improved Generalization**
   - Shared knowledge across tasks
   - Regularization effect

2. **Data Efficiency**
   - Learn from multiple datasets
   - Transfer knowledge between tasks

3. **Faster Training**
   - Shared feature extraction
   - One model instead of many

### Architecture Patterns

**Hard Parameter Sharing:**
- Shared hidden layers
- Task-specific output layers
- Most common approach

**Soft Parameter Sharing:**
- Separate models per task
- Regularize to be similar
- More flexible
