## 1. Why Explainable AI?

### The Black Box Problem

**Deep learning models:**
- High accuracy
- Complex internal representations
- Hard to understand decisions
- "Black box" nature

**Need for Explainability:**
- Trust and transparency
- Regulatory compliance (GDPR, etc.)
- Debugging and improvement
- Fairness and bias detection
- Safety-critical applications

### Types of Interpretability

**1. Global Interpretability:**
- Understand overall model behavior
- Feature importance
- Model-wide patterns

**2. Local Interpretability:**
- Explain individual predictions
- Why this specific output?
- Instance-level explanations

### Interpretability Methods

1. **Model-Specific:** Designed for specific architectures
2. **Model-Agnostic:** Work with any model
3. **Post-hoc:** Applied after training
4. **Intrinsic:** Built into model design
